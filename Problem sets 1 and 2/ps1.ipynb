```r
ceosal2 <- read_dta("data_ps1/CEOSAL2.dta")
```

```r
colnames(ceosal2)
```

```r
summary(ceosal2)
```

```r
lsalary <- log(ceosal2$salary)
lsales <- log(ceosal2$sales)
lmktval <- log(ceosal2$mktval)
```

```r
m <- lm(lsalary ~ lsales + lmktval, data=ceosal2)
summary(m)
```

The constant elasticity equation is 
$$
\log(salary) = 4.62 + .162\log(sales) + .107\log(mktval)
$$
$n=177$, $R^2=0.299$

The variables (sales and market value) explain about 29.9% of the variation in CEO salaries. The rest (70.1%) is due to other factors not in the model, randomness, measurement error, etc.

An increase of 1% in sales is associated with an increase of 0.162% in salary, ceteris paribus.

```r
m2 <- lm(lsalary ~ lsales + lmktval + profits, data=ceosal2)
summary(m2)
```

We cannot include profits in logarithmic form because profits are negative for 
nine of the companies in the sample. There is a very slight increase to the $R^2$, thus it is safe to assume that the addition of profits does not improve the model significantly.

```r
m3 <- lm(lsalary ~ lsales + lmktval + profits + ceoten, data=ceosal2)
summary(m3)
```

One more year as CEO increases predicted salary by 1.2%.

```r
cor(ceosal2$lmktval,ceosal2$profits)
```

The sample correlation between log(mktval) and profits is about .78, which is 
fairly high.  As we know, this causes no bias in the OLS estimators, although it can 
cause their variances to be large.  Given the fairly substantial correlation between 
market value and firm profits, it is not too surprising that the latter adds nothing to 
explaining CEO salaries.  Also, profits is a short term measure of how the firm is 
doing while mktval is based on past, current, and expected future profitability.
