```r
library(haven)
wage2 <- read_dta("data_ps3/WAGE2.DTA")
```

```r
dim(wage2)
colnames(wage2)
```

```r
#lecture's model (2)
reg1<-lm(lwage ~ educ + exper + tenure + married + south + urban + black + IQ, data=wage2)
summary(reg1)
```

```r
# use KWW instead of IQ (as a proxy for ability)
reg2<-lm(lwage ~ educ + exper + tenure + married + south + urban + black + KWW, data=wage2)
summary(reg2)
```

note the similarity in R^2, i.e. how much of the variation in lwage is explained by the models, and in the ccoefficient in IQ (0.0035591) and KWW (0.005028).
